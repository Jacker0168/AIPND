* **Question 2**  
Given an arbitrary ransom note string and another string containing letters from all the magazines, write a function that will return true if the ransom note can be constructed from the magazines ; otherwise, it will return false. Each letter in the magazine string can only be used once in your ransom note.  
***Note:***
You may assume that both strings contain only lowercase letters.  
canConstruct("a", "b") -> false  
canConstruct("aa", "ab") -> false  
canConstruct("aa", "aab") -> true  
canConstruct("aa", "aba") -> true  
Subscribe to see which companies asked this question.  
**Mark**  
1. collections.Counter
2. itertools.permutations

```python
# lc 383
# method 1 collections.Counter
import collections
def canConstruct1(ransom, magazine):
    return not collections.Counter(ransom) - collections.Counter(magazine)
print 'm1:', canConstruct('aa', 'aba')

# method 2
import itertools
def canConstruct2(ransom, magazine):
    for i in itertools.permutations(magazine, len(ransom)):
        if ransom == ''.join(i):
            return True
    return False
print 'm2:', canConstruct('aa', 'aba')

# method 3
import itertools
def canConstruct3(ransom, magazine):
    return ransom in [''.join(i) for i in itertools.permutations(magazine, len(ransom))]
print 'm2:', canConstruct('aa', 'aba')
```

* **Question 4**  
Given two strings s and t which consist of only lowercase letters. String t is generated by random shuffling string s and then add one more letter at a random position. Find the letter that was added in t.  
**Example:**  
Input:
s = "abcd"
t = "abcde"  
Output: e  
Explanation: 'e' is the letter that was added.  
**Mark**  
* string.lower()
* operator.xor
* ord

```python
# lc389
# method1
def findTheDifference1(s, t):
    s = s.lower()
    t = t.lower()
    return set(t) - set(s)
print findTheDifference1('abc', 'abcd')

# method2
import operator
def findTheDifference2(s, t):
    s = s.lower()
    t = t.lower()
    return chr(reduce(operator.xor, map(ord, (s + t))))
print findTheDifference2('abc', 'abcd')
```

```python
# lc409
# method 1
def longestPalindrome1(s):
    t = collections.Counter(s)
    return sum([t[x] for x in t if t[x] % 2 == 0]) + sum([t[x] - 1 for x in t if  t[x] % 2 == 1])\
                + max([1 for x in t if t[x] % 2 == 1] or [0])
# method 2
def longestPalindrome2(s):
    single = 0
    double = 0
    mix = 0
    for each in set(s):
        if s.count(each)%2 == 0:
            double += s.count(each)
        else:
            single = 1
            if s.count(each)/2 != 0:
                mix += (s.count(each)/2)*2
    return single + double + mix

# method 3
def longestPalindrome3(s):
    single = 1 if 1 in collections.Counter('abccccdd').values() else 0
    mixdouble = sum([s.count(each) if s.count(each)%2 == 0 else (s.count(each)/2)*2 for each in set(s)])
    return single + mixdouble
```

* **Question 8\***  
Given a string s and a non-empty string p, find all the start indices of p's anagrams in s. Strings consists of lowercase English letters only and the length of both strings s and p will not be larger than 20,100. The order of output does not matter.
**Example 1:**  
Input:  
s: "cbaebabacd" p: "abc"  
Output:  
[0, 6]  
**Explanation:**
The substring with start index = 0 is "cba", which is an anagram of "abc".
The substring with start index = 6 is "bac", which is an anagram of "abc".
Example 2:  
Input:  
s: "abab" p: "ab"  
Output:  
[0, 1, 2]  
**Explanation:**
The substring with start index = 0 is "ab", which is an anagram of "ab".
The substring with start index = 1 is "ba", which is an anagram of "ab".
The substring with start index = 2 is "ab", which is an anagram of "ab".

```python
import collections
# method 1
def findAnagrams(s, p):
    res = []
    pCounter = collections.Counter(p)
    sCounter = collections.Counter(s[:len(p)-1])
    for i in range(len(p)-1,len(s)):
        sCounter[s[i]] += 1   # include a new char in the window
        if sCounter == pCounter:    # This step is O(1), since there are at most 26 English letters
            res.append(i-len(p)+1)   # append the starting index
        sCounter[s[i-len(p)+1]] -= 1   # decrease the count of oldest char in the window
        if sCounter[s[i-len(p)+1]] == 0:
            del sCounter[s[i-len(p)+1]]   # remove the count if it is 0
    return res
print findAnagrams("abab", "ab")

# method 2
def findAnagrams(s, p):
    res = []
    left = right = 0
    count = len(p)
    dic = dict()
    for i in p:
        dic[i] = dic.get(i, 0) + 1
    while right < len(s):
        if s[right] in dic.keys():
            if dic[s[right]] >= 1:
                count = count - 1
            dic[s[right]] = dic[s[right]] - 1
        right = right + 1

        if count == 0:
            res.append(left)
        if right - left == len(p):
            if s[left] in dic.keys():
                if dic[s[left]] >= 0:
                    count = count + 1
                dic[s[left]] += 1
            left = left + 1
    return res
print findAnagrams("abab", "ab")
```

**Example 3:**  
Input: "abcabcabcabc"  
Output: True  
Explanation: It's the substring "abc" four times. (And the substring "abcabc" twice.)  


```python
# method 1
def repSubstring1(s):
    for i in range(1, len(s) / 2 + 1):
        if len(s) % i == 0:
            leng = len(s) / i
            if s[:i] * leng == s:
                return True
    return False

# method 2
def repSubstring2(s):
    ss = (s + s)[1:-1]
    return ss.find(s) != -1
print repSubstring('abab')
```

* **Question 11**  
Given a string, you need to reverse the order of characters in each word within a sentence while still preserving whitespace and initial word order.  
**Example 1:**  
Input: "Let's take LeetCode contest"  
Output: "s'teL ekat edoCteeL tsetnoc"  
**Note:** In the string, each word is separated by single space and there will not be any extra space in the string.  


```python
def reverseStr(s):
    return ' '.join([x[::-1] for x in s.split()])
print reverseStr("Let's take LeetCode contest")
```
